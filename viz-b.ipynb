```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris.describe()
```

```python
iris.info()
```

```python
sns.scatterplot(x='sepal_length', y ='petal_length' , data = iris , hue = 'species')
```

* We can notice that
1. The iris flower with a sepal length < 6 and petal length > 2 is most likely of type **setosa**.
2. The iris flower with petal length between 2.5 and 5 is most likely of type **versicolor**. 
3. The iris flower with petal length > 5 are most likely of type **virginica**.

```python
sns.set_style("whitegrid");
sns.pairplot(iris,hue="species");
```

```python
sns.boxplot(x="species",y="petal_length",data=iris)
```

* We can notice that we can seperate flowers using petal lenght while we can't do this using the sepal length.

```python
corr = iris.corr()
sns.heatmap(corr )
```

* The sepal length has a positive relation with petal length and petal width. while sepal width has a negative relation with petal length and petal width.
